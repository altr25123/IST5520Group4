<h3>Literature Review: </h3>
There exist several reviews investigating the relationship between city hotspot centers and their contribution to the growing crime rate. In 2004, a research team from the University of Maryland investigated crime incidents for 14 years at the level of street segments in Seattle. They found out consistently on a year-by-year basis that over 50 percent of crime in the City of Seattle was significantly concentrated in approximately 4.5 percent of street segments (Weisburd et al. 2004). These analyses revealed high crime stability in city hotspot centers over time. Furthermore, Braga et.al 2011 investigated the impact of hotspots on crime in the city of Boston. They highlighted that crime rates are highly concentrated in micro places and hotspots. Their finding suggests the need for concentrated efforts on specific places (hotspots) rather than increasing policing efforts for the whole city (Braga, Hureau, and Papachristos 2011). Several studied theories show that increased crime rates in communities are heavily linked to more alcohol outlets in these hotspot centers. Research confirms the complex relationship and association between alcohol outlet density and violent crimes. Also, further observations predict neighborhoods with increasing cases of violent crime rates were observed to be of lower income and in areas with less guardianship (Lipton, Yang, Braga, Goldstick, Newton, Rura 2012).

<h2>Data Manipulation:</h2>
Upon importing 2015-2022 Boston crime incident reports into Python we found that our data was mainly qualitative or categorical variables. As we began cleansing the data we came across a lot of issues that would affect how we would use the data for visualization. The major issue though was in the 2015-2018 data they used OFFENSE_CODE_GROUP and UCR_PART but they did not in 2019-2022 and instead entered ‘None’ into the dataset. Due to this issus we decided the best course of action was to delete both of the columns from our dataset. The next issue that occurred was that in 2015-2018 data in the SHOOTING column they used Y to indicate there had been a shooting and NONE to indicate no shooting but in the 2019-2022 data they used 1 for yes and 0 for no. We had to reconfigure the entire SHOOTING column for 2015-2018 data to replace the Y and NONE with 1 and 0. We then went in to the Location column and got rid of the ‘(‘ and ‘)’ around the latitude and longitude. 

The next issue that we ran into while manipulating the data was most of the data was qualitative data and wasn’t interval or ratio data. This made it hard to normalize the data and do principal component analysis as there were no numerical data types. We then created multiple frequency tables to evaluate the data. We found this to be the easiest method to analyze the data because it provided us with counts of each qualitative value. Below there are two frequency tables. One is OFFENSE_CODE and YEAR which shows us how many of a certain type of offence occurs in a particular year. The other frequency table shows how many certain types of offenses occur based on offense code occur at a certain hour of the day. As shown below is the data manipulation and cleansing that we did to prepare our data. 


<h3>Importing Data</h3>

```python
# Import modules
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA

# Configure inline mode
%matplotlib inline 
```

```python
dat2015 = pd.read_csv("crime-incident-reports-2015.csv")
dat2016 = pd.read_csv("crime-incident-reports-2016.csv")
dat2017 = pd.read_csv("crime-incident-reports-2017.csv")
dat2018 = pd.read_csv("crime-incident-reports-2018.csv")
dat2019 = pd.read_csv("crime-incident-reports-2019.csv")
dat2020 = pd.read_csv("crime-incident-reports-2020.csv")
dat2021 = pd.read_csv("crime-incident-reports-2021.csv")
dat2022 = pd.read_csv("crime-incident-reports-2022.csv")
```

<h3>Creating Consistency Between Years</h3>

```python
dat2015['SHOOTING']=dat2015['SHOOTING'].map({'Y': 1, 'None': 0})
dat2015['SHOOTING']=dat2015['SHOOTING'].fillna(0)
dat2015['Lat']=dat2015['Lat'].fillna(0)
dat2015['Long']=dat2015['Long'].fillna(0)
dat2015['SHOOTING']=dat2015['SHOOTING'].astype('int')
dat2016['SHOOTING']=dat2016['SHOOTING'].map({'Y': 1, 'None': 0})
dat2016['SHOOTING']=dat2016['SHOOTING'].fillna(0)
dat2016['Lat']=dat2016['Lat'].fillna(0)
dat2016['Long']=dat2016['Long'].fillna(0)
dat2016['SHOOTING']=dat2016['SHOOTING'].astype('int')
dat2017['SHOOTING']=dat2017['SHOOTING'].map({'Y': 1, 'None': 0})
dat2017['SHOOTING']=dat2017['SHOOTING'].fillna(0)
dat2017['Lat']=dat2017['Lat'].fillna(0)
dat2017['Long']=dat2017['Long'].fillna(0)
dat2017['SHOOTING']=dat2017['SHOOTING'].astype('int')
dat2018['SHOOTING']=dat2018['SHOOTING'].map({'Y': 1, 'None': 0})
dat2018['SHOOTING']=dat2018['SHOOTING'].fillna(0)
dat2018['Lat']=dat2018['Lat'].fillna(0)
dat2018['Long']=dat2018['Long'].fillna(0)
dat2018['SHOOTING']=dat2018['SHOOTING'].astype('int')
```

<h3>Combining Datasets</h3>

```python
#Combine all datasets
CrimeAll=[dat2015,dat2016,dat2017,dat2018,dat2019,dat2020,dat2021,dat2022]
CrimeAll
CrimeAll = pd.concat(CrimeAll)
CrimeAll
```

<h3>Cleaning Location Data</h3>

```python
CrimeAll['Location'] = CrimeAll['Location'].str.replace('(','')
CrimeAll['Location'] = CrimeAll['Location'].str.replace(')','')

CrimeAll.head()
```

<h3>Dropping Columns</h3>

```python
 CrimeAll=CrimeAll.drop(['OFFENSE_CODE_GROUP', 'UCR_PART'], axis=1)
CrimeAll
```
