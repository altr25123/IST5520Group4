<h3>Introduction: </h3>
Crime has proven to significantly influence the life pattern of people in a given community. Community life, people’s choice of work, living and shopping are actively impacted by this canker. Many research confirms that the rate of crime is highly centered in special or “hotspot” areas in cities. These active places have been reported to be the primary sources from where crimes spread throughout an entire city. Herein, we investigated the effects of these “hotspot” centers on the spread of crime in the city of Boston. Data from the Boston Police Department was collected and employed to develop a predictive model that evaluates the relationship between various forms of crime and special locations in the city of Boston. The work is highly significant as it highlights the relevance of places in the decision making processes of robbers and other reported crimes. Knowledge of this information is bound to help all in the crime-fighting agencies including other stakeholders such as businessmen, landlords, and other employers with necessary information needed to combat and prevent crime in the city of Boston.  This will go a long way to improve the culture and living of the people in the city. 

<h3>Literature Review: </h3>
There exist several reviews investigating the relationship between city hotspot centers and their contribution to the growing crime rate. In 2004, a research team from the University of Maryland investigated crime incidents for 14 years at the level of street segments in Seattle. They found out consistently on a year-by-year basis that over 50 percent of crime in the City of Seattle was significantly concentrated in approximately 4.5 percent of street segments (Weisburd et al. 2004). These analyses revealed high crime stability in city hotspot centers over time. Furthermore, Braga et.al 2011 investigated the impact of hotspots on crime in the city of Boston. They highlighted that crime rates are highly concentrated in micro places and hotspots. Their finding suggests the need for concentrated efforts on specific places (hotspots) rather than increasing policing efforts for the whole city (Braga, Hureau, and Papachristos 2011). Several studied theories show that increased crime rates in communities are heavily linked to more alcohol outlets in these hotspot centers. Research confirms the complex relationship and association between alcohol outlet density and violent crimes. Also, further observations predict neighborhoods with increasing cases of violent crime rates were observed to be of lower income and in areas with less guardianship (Lipton, Yang, Braga, Goldstick, Newton, Rura 2012).

<h3>Research Questions: </h3>
There are several problems that we are hoping to address with this particular dataset on crime incidents in the Boston area. A question we would like to answer is if crime is increasing in the Boston area. From there, we would like to better understand the types of crimes that are occuring in various areas across the city based on their district, street, even as detailed as their exact geographical location using latitude and longitude data. Once we have a better understanding of what the data looks like and is telling us, we would like to use this as a way to make inferences based on the data as to when and where we can expect crimes as well as the offense type of said crime. We are also interested in looking into if there is a correlation between the incident location (place where the crime physically took place) and the reporting area (area where crime was reported from). With this model we would, in layman's terms, be able to “predict” crime, although we are really just using the data to say “Based on this previous data, we can infer/expect crime to increase/decrease this much over this amount of time and here is what types of crimes that involves. Along with the aforementioned, we hope to shed light on specific areas within Boston that have higher crime occurrences that may need more resources than other areas in the city in the future.

These reports are provided by the Boston Police Department of incidents that their officers attended to. These reports focus heavily on the location and time that a crime incident has occured. Based on the columns above the crime incidents dataset deems appropriate for helping address our research questions about being able to predict locations of concern where crime incidents may be more likely to occur. The crime incidents reports include columns about the district, reporting area, street address, latitude, longitude, and location. The ability of having the exact location down to the latitude and longitude will allow our group to analyze very specific micro regions and allow the possibility of having a very accurate prediction of where crimes will occur. On top of having an accurate location the reports also have very precise data and times included. Columns such as occured_on_date, year, month, day_of_the_week, and hour will support our ability to predict at what days and times these crimes are occurring in these distinct locations. The reports allow us to possibly predict down to the minute where a crime may occur. Thus the crime incidents dataset provides an ideal source for helping address our research questions pertaining to where and when crime incidents are occuring in Boston.
